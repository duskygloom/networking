```python
import matplotlib.pyplot as plot
from math import cos, pi
from numpy import arange

A   = 1
f   = 4
phi = 0

nsamples = 50

stream  = "111110101100"

nsignals    = len(stream)/3
xstep       = 1
title       = f"8-PSK ({stream})"
```

```python
fig, axis = plot.subplots()

fig.set_figwidth(nsignals/xstep)
fig.set_figheight(3)
axis.set_title(title)
axis.set_ylabel("Signal")
axis.set_yticks((-2, -1, 0, 1, 2))
axis.set_ylim(-2.5, 2.5)
axis.set_xlabel("Time")
xticks = arange(0, nsignals+xstep, xstep)
axis.set_xticks(xticks)
axis.set_xlim(-xstep, nsignals+xstep)
axis.set_xticklabels([round(i, 4) for i in xticks])
axis.plot([0, nsignals], [0, 0], linestyle="dashed", color="gray")

x = []
y = []

base = 0

# Binary ASK
# for bit in stream:
#     for sample in range(nsamples):
#         t = base + sample/nsamples
#         x.append(t)
#         y.append(int(bit)*cos(2*pi*f*t+phi))
#     base += 1

# Binary FSK
# for bit in stream:
#     for sample in range(nsamples):
#         t = base + sample/nsamples
#         x.append(t)
#         y.append(A*cos(2*pi*2*(int(bit)+1)*t+phi))
#     base += 1

# Binary PSK
# for bit in stream:
#     for sample in range(nsamples):
#         t = base + sample/nsamples
#         x.append(t)
#         phase = int(bit) * pi
#         y.append(A*cos(2*pi*f*t+phase))
#     base += 1

# QPSK
# for i in range(0, len(stream), 2):
#     bits    = stream[i:i+2]
#     bit     = -1
#     if bits == "00":
#         bit = 0
#     elif bits == "01":
#         bit = 1
#     elif bits == "11":
#         bit = 2
#     elif bits == "10":
#         bit = 3
#     for sample in range(nsamples):
#         t = base + sample/nsamples
#         x.append(t)
#         phase = (2*bit+1)*pi/4
#         y.append(A*cos(2*pi*f*t+phase))
#     base += 1

# 8-PSK
for i in range(0, len(stream), 3):
    bits    = stream[i:i+3]
    bit     = 0
    if bits == "000":
        bit = 0
    elif bits == "001":
        bit = 1
    elif bits == "011":
        bit = 2
    elif bits == "010":
        bit = 3
    elif bits == "110":
        bit = 4
    elif bits == "111":
        bit = 5
    elif bits == "101":
        bit = 6
    elif bits == "100":
        bit = 7
    for sample in range(nsamples):
        t = base + sample/nsamples
        x.append(t)
        phase = (2*bit+1)*pi/8
        y.append(A*cos(2*pi*f*t+phase))
    base += 1

axis.plot(x, y)
```
